```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Khởi tạo cơ bản
#Tạo 1 fig và axes duy nhất
fig, ax = plt.subplots()
#Tạo 1 fig và 2x2 axes
fig, axs = plt.subplots(2,2, figsize=(10,8))
plt.show()
```

```python
# Các hàm vẽ cơ bản
x = np.linspace(0, 2*np.pi, 200)
y_sin = np.sin(x)
y_cos = np.cos(x)
fig, axs = plt.subplots(2,1)
# Biểu đồ đường (Line Plot)
axs[0].plot(x, y_sin)
axs[1].plot(x, y_cos,
              color='r',                #'r','g','b'
              linestyle='--',           #'-', ':', '--', '-.'
              linewidth='2',
              marker='o',               #'.', 'o', '*'
              markersize=4,
              label='Cosine curve')
plt.show()
```

```python
# Vẽ biểu đồ phân tán (Scatter Plot)
rand_x = np.random.rand(50)
rand_y = np.random.rand(50)
sizes = 100 * np.random.rand(50)
colors = np.random.rand(50)
fig, ax = plt.subplots()
ax.scatter(rand_x, rand_y,
           s=sizes,                # Mảng kích thước cho từng điểm
           c=colors,               # Mảng màu cho từng điểm
           alpha=0.6)              # Độ trong suốt
plt.show()
```

```python
# Vẽ biểu đồ tần suất (Histogram)
data_hist = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(data_hist, 
        bins=30,                   # Số cột (khoảng)
        color='purple',
        edgecolor='black',         # Màu viền cột
        alpha=0.7)                 # Độ trong suốt
plt.show()
```

```python
categories = ['A', 'B', 'C']
values = [10, 25, 15]
fig, ax = plt.subplots(2,1)
# Vẽ biểu đồ cột (Bar chart)
ax[0].bar(categories, values, color='b', width=0.5)

# Vẽ biểu đồ cột ngang (Horizontal Bar Chart)
ax[1].barh(categories, values, color='salmon')
plt.show()
```

```python
data1 = np.random.normal(0, 1, 100)
data2 = np.random.normal(3, 2, 100)
data3 = np.random.normal(1, 1.5, 100)
all_data = [data1, data2, data3]
fig, ax = plt.subplots()
box = ax.boxplot(all_data,
                 vert=True,         # True: đứng, False: ngang
                 patch_artist=True, # Cho phép tô màu các hộp
                 labels=['Nhóm 1', 'Nhóm 2', 'Nhóm 3']) # Nhãn cho mỗi hộp
colors = ['lightblue', 'lightgreen', 'lightpink']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)
plt.show()
```

```python
sizes = [15, 30, 45, 10]
labels = ['Cáo', 'Chó', 'Mèo', 'Sư tử']
explode = (0, 0.1, 0, 0) # Tách "Chó" ra khỏi hình tròn
fig, ax = plt.subplots()
ax.pie(sizes,
       explode=explode,
       labels=labels,
       autopct='%1.1f%%',   # Hiển thị % với 1 chữ số thập phân
       shadow=True,
       startangle=90)      # Góc bắt đầu

ax.axis('equal')  # Đảm bảo biểu đồ là hình tròn
ax.set_title('Tỉ lệ các loài vật')
```

```python
# Tạo ma trận dữ liệu 2D
matrix_data = np.random.rand(10, 10) * 100
fig, ax = plt.subplots()
img = ax.imshow(matrix_data,
                cmap='hot',          # Bảng màu: 'viridis', 'plasma', 'inferno', 'magma', 'hot', 'coolwarm'
                interpolation='nearest') # Cách nội suy màu

# Thêm thanh màu
fig.colorbar(img, ax=ax, label='Cường độ')
plt.show()
```

```python
x_grid = np.linspace(-3.0, 3.0, 100)
y_grid = np.linspace(-3.0, 3.0, 100)
X, Y = np.meshgrid(x_grid, y_grid)
Z = np.sin(np.sqrt(X**2 + Y**2))
fig, ax = plt.subplots()
# ax.contour: Chỉ vẽ đường viền
contour_lines = ax.contour(X, Y, Z, colors='black')
ax.clabel(contour_lines, inline=True, fontsize=8) # Thêm nhãn giá trị lên đường viền

# ax.contourf: Tô màu vùng giữa các đường viền
contour_filled = ax.contourf(X, Y, Z, cmap='viridis')
fig.colorbar(contour_filled, ax=ax, label='Giá trị Z')
plt.show()
```

```python

fig, ax = plt.subplots()
x = np.linspace(0, 2*np.pi, 200)
y_sin = np.sin(x)
y_cos = np.cos(x)
ax.plot(x,y_sin,label='sin curve')
ax.plot(x,y_cos,label='cos curve')
# Tiêu đề và Nhãn (Titles & Labels)



ax.set_title('Tiêu đề của Subplot')
ax.set_xlabel('Nhãn trục X')
ax.set_ylabel('Nhãn trục Y')

# Chú thích (Legend)
ax.legend() 
ax.legend(loc='upper left') 

# Lưới (Grid)
ax.grid(True)                    # Bật lưới
ax.grid(color='gray', linestyle=':', linewidth=0.5, alpha=0.7)

# Giới hạn trục (Axis Limits)
ax.set_xlim(0, 10)               # Đặt giới hạn trục X từ 0 đến 10
ax.set_ylim(-1.5, 1.5)           # Đặt giới hạn trục Y từ -1.5 đến 1.5
ax.axis('tight')                 # Tự động co dãn trục cho vừa dữ liệu
ax.axis('equal')                 # Đảm bảo tỉ lệ 1:1 giữa X và Y

# Chỉnh Figure
fig.suptitle('Tiêu đề chính của toàn bộ hình vẽ', fontsize=20)
# Tự động căn chỉnh các subplot để không bị chồng chéo
fig.tight_layout()
plt.show()
```
